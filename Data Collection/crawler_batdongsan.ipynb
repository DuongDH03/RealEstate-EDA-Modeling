# Crawling

```python
def random_sleep(min_seconds, max_seconds):
    """Sleep for a random number of seconds between min and max."""
    time.sleep(random.uniform(min_seconds, max_seconds))
```

```python
def start(url):
    driver.get(url)
    driver.implicitly_wait(5)
    # Wait for the page to load
    time.sleep(3)

    # Fake human behavior by scrolling down
    for _ in range(2):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        random_sleep(min_sleep, max_sleep)

    starting_points = driver.find_elements(By.CSS_SELECTOR, ".re__card-info-content")
    if starting_points:
        starting_point = random.choice(starting_points)
    else:
        raise Exception("No starting points found")

    # Click using JavaScript
    driver.execute_script("arguments[0].click();", starting_point)

    random_sleep(min_sleep, max_sleep)
```

```python

start(url)
# Prepare data for JSON
datas = []
count = 0
impatience = 0
collected = 0

# Dynamically name the JSON file based on the current timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
target_filename = f"crawl_{timestamp}.jsonl"
# Create the target folder if it doesn't exist
if not os.path.exists(target_folder):
    os.makedirs(target_folder)
jsonl_filename = os.path.join(target_folder, target_filename)
try:
    while count < 200:
        data = []
        try :
            title = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, ".re__pr-title"))
            )
        except TimeoutException as e:
            print("Description not found, trying again...")
            next_points = driver.find_elements(By.CSS_SELECTOR, ".re__card-info-content")
            if next_points:
                next_point = random.choice(next_points)
                driver.execute_script("arguments[0].click();", next_point)
                print("Next point found")
                count = 0
            else:
                print("Next point not found")
                count += 1
                impatience += 1
                if impatience >= 1:
                    try:
                        return_point = driver.find_element(By.CSS_SELECTOR, ".lcUWFe+ .kBsqiG .bsIpGG")
                        driver.execute_script("arguments[0].click();", return_point)
                        print("Return point found")
                        impatience = 0
                    except Exception as e:
                        try:
                            print("Return point from login not found")
                            return_point = driver.find_element(By.CSS_SELECTOR, ".tablet+ .lv0 .text")
                            driver.execute_script("arguments[0].click();", return_point)
                            print("Return point from projects found")
                        except Exception as e:
                            print("Page 404 not found, restarting...")
                            start(url)
                            continue
            continue
        # driver.execute_script("arguments[0].scrollIntoView();", next_point)
        print("Description and title found")

        labels = driver.find_elements(By.CSS_SELECTOR, ".re__pr-specs-content-item-title")
        elements = driver.find_elements(By.CSS_SELECTOR, ".re__pr-specs-content-item-value")
        if labels and elements:
            print("Labels and elements found")
            data.append({"title": title.text})
            data.append({"url": driver.current_url})
        for label, element in zip(labels, elements):
            if label and element:
                data.append({label.text: element.text})

        # Write the entire datas list to a JSON file
        with open(jsonl_filename, "a", encoding="utf-8") as jsonl_file:
            jsonl_file.write(json.dumps(data, ensure_ascii=False) + "\n")
        collected += 1
        if collected >= MAX_CRAWL:
            print("Reached maximum crawl limit.")
            break
        random_sleep(min_sleep, max_sleep)
        # Similar listing recommendations
        #anchors = driver.find_elements(By.CSS_SELECTOR, ".re__product-keyword-box .re__sidebar-box-item:nth-child(1) .re__link-se")
        anchors = driver.find_elements(By.CSS_SELECTOR, ".re__product-keyword-box :nth-child(1)")
        
        # driver.execute_script("arguments[0].scrollIntoView();", next_point)
        if anchors:
            print("Anchor found")
            anchor = random.choice(anchors)
        else:
            print("Anchor not found")
            count += 1
            continue
        driver.execute_script("arguments[0].click();", anchor)

        random_sleep(min_sleep, max_sleep)
        next_points = driver.find_elements(By.CSS_SELECTOR, ".re__card-info-content")
        if next_points:
            next_point = random.choice(next_points)
            driver.execute_script("arguments[0].click();", next_point)
            print("Next point found")
            count = 0
        else:
            print("Next point not found")
            count += 1
        continue
finally:
    driver.quit()
```

# Flatten the output json file

```python
import json
import os

input_folder = "../data/json_batdongsan"

for filename in os.listdir(input_folder):
    if filename.endswith(".jsonl"):
        if(filename.endswith("_flattened.jsonl")):
            print(f"File {filename} already flattened. Skipping...")
            continue
        input_file = os.path.join(input_folder, filename)
        output_file = os.path.join(input_folder, filename.replace(".jsonl", "_flattened.jsonl"))

        # Check if the output file already exists
        if os.path.exists(output_file):
            print(f"Output file {output_file} already exists. Skipping...")
            continue

    # Open the input and output files
    try:
        with open(input_file, "r", encoding="utf-8") as infile, open(output_file, "w", encoding="utf-8") as outfile:
            for line in infile:
                try:
                    # Parse the nested JSON object
                    nested_data = json.loads(line)
                    # Flatten the nested structure
                    flat_data = {}
                    for item in nested_data:
                        flat_data.update(item)
                    
                    # Write the flattened data to the output file
                    outfile.write(json.dumps(flat_data, ensure_ascii=False) + "\n")
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON: {e}, skipping line.")
                    continue
            print(f"Flattened data written to {output_file}")
    except FileNotFoundError as e:
        print(f"Error processing file {input_file}: {e}")
        continue
    if os.path.exists(input_file) and os.path.exists(output_file):
        os.remove(input_file)
        print(f"Removed input file {input_file}")
```

# Remove duplicates in csv file

```python
# Path to the input CSV file
input_csv = "../data/raw_batdongsan/batdongsan.csv"

# Path to the output CSV file
output_csv = "../data/raw_batdongsan/unique_batdongsan.csv"

# Read the CSV file and remove duplicates
unique_rows = set()
with open(input_csv, "r", encoding="utf-8") as infile, open(output_csv, "w", encoding="utf-8") as outfile:
    header = infile.readline()
    outfile.write(header)  # Write the header to the output file
    for line in infile:
        if line not in unique_rows:
            unique_rows.add(line)
            outfile.write(line)

print(f"Duplicates removed. Deduplicated file saved to {output_csv}")
```
